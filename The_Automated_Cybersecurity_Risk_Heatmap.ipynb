```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. Create a synthetic GRC Risk Dataset
data = {
    'Risk_ID': ['R1', 'R2', 'R3', 'R4', 'R5', 'R6', 'R7', 'R8'],
    'Threat': ['Ransomware', 'Data Leak', 'Phishing', 'Misconfiguration', 'Insider Threat', 'Zero Day', 'API Breach', 'SQL Injection'],
    'Likelihood': [4, 2, 5, 3, 2, 1, 4, 3], # Scale 1-5
    'Impact': [5, 4, 3, 2, 3, 5, 4, 4]      # Scale 1-5
}

df = pd.DataFrame(data)

# 2. Create a 5x5 matrix for the heatmap
heatmap_data = np.zeros((5, 5))
for index, row in df.iterrows():
    # Subtract 1 because Python indices start at 0
    heatmap_data[row['Impact']-1, row['Likelihood']-1] += 1

# 3. Visualization
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap='YlOrRd',
            xticklabels=[1,2,3,4,5], yticklabels=[1,2,3,4,5])

plt.title('Cybersecurity Risk Heatmap', fontsize=15)
plt.xlabel('Likelihood', fontsize=12)
plt.ylabel('Impact', fontsize=12)
plt.gca().invert_yaxis() # Highest impact at the top
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Your GRC Risk Data
data = {
    'Threat': ['Ransomware', 'Data Leak', 'Phishing', 'Misconfig', 'Insider', 'Zero Day', 'API Breach', 'SQL Inj'],
    'Likelihood': [4, 2, 5, 3, 2, 1, 4, 3],
    'Impact': [5, 4, 3, 2, 3, 5, 4, 4]
}
df = pd.DataFrame(data)

# 2. Create the visualization
plt.figure(figsize=(10, 8))
# We use a scatter plot over the heatmap to place names accurately
sns.heatmap(np.zeros((5, 5)), cmap='YlOrRd', cbar=False,
            xticklabels=[1,2,3,4,5], yticklabels=[1,2,3,4,5], alpha=0.3)

# 3. Add the labels for each threat
for i in range(df.shape[0]):
    plt.text(df.Likelihood[i]-0.5, df.Impact[i]-0.5, df.Threat[i],
             ha='center', va='center', fontweight='bold', color='darkred')

plt.title('Annotated Cybersecurity Risk Heatmap', fontsize=15)
plt.xlabel('Likelihood', fontsize=12)
plt.ylabel('Impact', fontsize=12)
plt.gca().invert_yaxis()
plt.show()
```
